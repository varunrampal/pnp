# Remove Duplicate Plant Names from JSON
This notebook loads the plant data, identifies duplicate `Name` values, keeps the first occurrence, and writes a cleaned JSON file.

```python
import json
from pathlib import Path
```

```python
data_path = Path('src/json/PlantsList.json')
with data_path.open('r', encoding='utf-8') as f:
    plants = json.load(f)

len(plants)
```

```python
seen = set()
duplicates = []
for item in plants:
    name = (item.get('Name') or '').strip()
    if name in seen:
        duplicates.append(name)
    else:
        seen.add(name)

len(duplicates)
```

```python
seen = set()
deduped = []
for item in plants:
    name = (item.get('Name') or '').strip()
    if name not in seen:
        seen.add(name)
        deduped.append(item)

output_path = Path('src/json/PlantsList_Deduped.json')
with output_path.open('w', encoding='utf-8') as f:
    json.dump(deduped, f, indent=4, ensure_ascii=False)

print(f'Total original entries: {len(plants)}')
print(f'Total deduplicated entries: {len(deduped)}')
print(f'Duplicates removed: {len(plants) - len(deduped)}')
print(f'Wrote cleaned JSON to: {output_path}')
```
